```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
datos = pd.read_csv('games.csv', index_col="Id")
datos.head()
```

```python
datos.info()
```

```python
datos.isnull().any()
```

```python
filas_nulas = datos[datos.isnull().any(axis = 1)]
print(filas_nulas)
```

```python
sin_nulos = datos.dropna()
sin_nulos.info()
```

```python
def filtro_rpg(genres):
    return 'RPG' in genres

datosRPG = sin_nulos[sin_nulos['Genres'].apply(filtro_rpg)]
datosRPG
```

```python
nomresRPG = datosRPG[['Title','Rating']]
nomresRPG.head(10)
```
